### <font color="blue">[1.1)](#toa)(10 pts) You have a contract to create a **single**, suitable data structure to hold all the table information below so that data is *only* accessible by empirical formula. Also, *the data structure must allow for an automatic sum of the mole fractions (see 1.2) below)*. Explain the design of your data structure.</font><a id="a11"></a>

| Species | Empirical Formula | Mole Fraction | Value [%] |
|-------------|-----------------|--------------|-------|
|ethyl nitrite     | C2H5ONO | $x_\text{C2H5ONO}$ | 90   |
|hydroxyethyl      | C2H5O   | $x_\text{C2H5O}$   | 1  |
|nitrogen monoxide | NO      | $x_\text{NO}$      | 1   |
|methyl radical    | CH3     | $x_\text{CH3}$     | 0.5  |
|formaldehyde      | CH2O    | $x_\text{CH2O}$    | 0.5     |
|acetaldehyde      | CH3CHO  | $x_\text{CH3CHO}$  | 1  |
|nitroxyl          | HNO     | $x_\text{HNO}$     | 0.5      |
|ethanol           | C2H5OH  | $x_\text{C2H5OH}$  | 1      |
|nitrogen hydroxide| NOH     | $x_\text{NOH}$     | 0.5      |
|nitric oxide      | N2O     | $x_\text{N2O}$     | 1      |
|water             | H2O     | $x_\text{H2O}$     | 1   |
|formamide         | CH3NO   | $x_\text{CH3NO}$   | 1      |
|formaldoxime      | CH2=NOH | $x_\text{CH2=NOH}$ | 0.5   |
|hydrogen cyanide  | HCN     | $x_\text{HCN}$     | 0.5      |
    

```python
import numpy as np
import pandas as pd

formula_search = 'NO' #searches by chemical formula

ethyl_nitrite     = {'species'          :'ethyl nitrite',
                     'empirical formula':'C2H5ONO',
                     'mole fraction'    :'x'+'C2H5ONO',
                     'Value[%]'         :90}
hydroxyethyl      = {'species'          :'hydroxyethyl',
                     'empirical formula':'C2H5O',
                     'mole fraction'    :'x'+'C2H5O',
                     'Value[%]'         :1}
nitrogen_monoxide = {'species'          :'nitrogen monoxide',
                     'empirical formula':'NO',
                     'mole fraction'    :'x'+'NO',
                     'Value[%]'         :1}
methyl_radical    = {'species'          :'methyl radical',
                     'empirical formula':'CH3',
                     'mole fraction'    :'x'+'CH3',
                     'Value[%]'         :0.5}
formaldehyde      = {'species'          :'formaldehyde',
                     'empirical formula':'CH2O',
                     'mole fraction'    :'x'+'CH2O',
                     'Value[%]'         :0.5}
acetaldehyde      = {'species'          :'acetaldehyde',
                     'empirical formula':'CH2O',
                     'mole fraction'    :'x'+'CH2O',
                     'Value[%]'         :1}
nitroxyl          = {'species'          :'nitroxyl',
                     'empirical formula':'HNO',
                     'mole fraction'    :'x'+'HNO',
                     'Value[%]'         :0.5}
ethanol           = {'species'          :'ethanol',
                     'empirical formula':'C2H5OH',
                     'mole fraction'    :'x'+'C2H5OH',
                     'Value[%]'         :1}
nitrogen_hydroxide= {'species'          :'nitrogen hydroxide',
                     'empirical formula':'NOH',
                     'mole fraction'    :'x'+'NOH',
                     'Value[%]'         :0.5}
nitric_oxide      = {'species'          :'nitric oxide',
                     'empirical formula':'N2O',
                     'mole fraction'    :'x'+'N2O',
                     'Value[%]'         :1}
water             = {'species'          :'water',
                     'empirical formula':'H2O',
                     'mole fraction'    :'x'+'H2O',
                     'Value[%]'         :1}
formamide         = {'species'          :'formamide',
                     'empirical formula':'CH3NO',
                     'mole fraction'    :'x'+'CH3NO',
                     'Value[%]'         :1}
formaldoxime      = {'species'          :'formaldoxime',
                     'empirical formula':'CH2=NOH',
                     'mole fraction'    :'x'+'CH2=NOH',
                     'Value[%]'         :0.5}
hydrogen_cyanide  = {'species'          :'hydrogen cyanide',
                     'empirical formula':'HCN',
                     'mole fraction'    :'x'+'HCN',
                     'Value[%]'         :0.5}
periodic_table = {'ethyl nitrite'     :ethyl_nitrite, 
                  'hydroxyethyl'      :hydroxyethyl,
                  'nitrogen_monoxide' :nitrogen_monoxide,
                  'methyl radical'    :methyl_radical,
                  'formaldehyde'      :formaldehyde,
                  'acetaldehyde'      :acetaldehyde,
                  'nitroxyl'          :nitroxyl,
                  'ethanol'           :ethanol,
                  'nitrogen hydroxide':nitrogen_hydroxide,
                  'nitric_oxide'      :nitric_oxide,
                  'formamide'         :formamide,
                  'formaldoxime'      :formaldoxime,
                  'water'             :water,
                  'hydrogen cyandide' :hydrogen_cyanide
                 }
for key in periodic_table.keys():
    element = periodic_table[key]
    if formula_search == element['empirical formula']:
        print('you searched for:    ', formula_search)
        print('the species is:      ', element['species'])
        print('the formula is:      ', element['empirical formula'])
        print('the mole fraction is:', element['mole fraction'])
        print('the value% is:       ', element['Value[%]'])
        
df = pd.DataFrame(data=[ethyl_nitrite, 
                        hydroxyethyl,
                        nitrogen_monoxide,
                        methyl_radical,
                        formaldehyde,
                        acetaldehyde,
                        nitroxyl,
                        ethanol,
                        nitrogen_hydroxide,
                        nitric_oxide,
                        water,
                        formamide,
                        formaldoxime,
                        hydrogen_cyanide
                       ])

    
print('')
print(df.to_string(index=False))
```
